## Convert to LJSpeech

```python
import pandas as pd
import shutil
from pathlib import Path

# Đường dẫn nguồn
src_metadata = "data/metadata.csv"
src_wavs_dir = Path("data/wavs")

# Đường dẫn đích
output_dir = Path("processed_data")
output_wavs_dir = output_dir / "wavs"

# Tạo thư mục
output_wavs_dir.mkdir(parents=True, exist_ok=True)

# Đọc metadata gốc
df = pd.read_csv(src_metadata, sep="|")

metadata_lines = []

for _, row in df.iterrows():
    wav_path = Path(row["file_path"])

    # Kiểm tra file tồn tại
    if not wav_path.exists():
        print(f"Warning: {wav_path} not found")
        continue

    file_id = wav_path.stem

    # Copy file wav
    shutil.copy2(
        wav_path,
        output_wavs_dir / wav_path.name
    )

    # Format LJSpeech:
    # <file_id>|<transcript>|<normalized_transcript>
    metadata_lines.append(
        f"{file_id}|{row['raw_transcript']}|{row['normalized_transcript']}"
    )

# Ghi metadata mới
with open(output_dir / "metadata.csv", "w", encoding="utf-8") as f:
    f.write("\n".join(metadata_lines))

print(f"Done! Converted {len(metadata_lines)} samples.")
print(f"Output: {output_dir.resolve()}")
```

## Take 20%

```python
import pandas as pd
import shutil
from pathlib import Path

# Nguồn
src_dir = Path("processed_data")
src_wavs = src_dir / "wavs"
src_metadata = src_dir / "metadata.csv"

# Đích
dst_dir = Path("processed_data_20pct")
dst_wavs = dst_dir / "wavs"

dst_wavs.mkdir(parents=True, exist_ok=True)

# Đọc metadata LJSpeech
df = pd.read_csv(
    src_metadata,
    sep="|",
    header=None,
    names=["file_id", "text", "normalized_text"]
)

# Lấy ngẫu nhiên 20%
df_20 = df.sample(frac=0.2, random_state=42)

# (Tùy chọn) sort lại theo file_id
try:
    df_20 = df_20.sort_values(
        by="file_id",
        key=lambda s: s.astype(int)
    )
except:
    df_20 = df_20.sort_values(by="file_id")

# Copy wav tương ứng
for file_id in df_20["file_id"]:
    src_file = src_wavs / f"{file_id}.wav"
    dst_file = dst_wavs / f"{file_id}.wav"

    if src_file.exists():
        shutil.copy2(src_file, dst_file)
    else:
        print(f"Missing: {src_file}")

# Ghi metadata mới
df_20.to_csv(
    dst_dir / "metadata.csv",
    sep="|",
    header=False,
    index=False
)

print(f"Selected {len(df_20)} samples")
print(f"Saved to {dst_dir}")
```
